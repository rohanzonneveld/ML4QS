```python
# plot gyroscope data
data = pd.read_csv('intermediate_datafiles/dataset.csv')

fig, axs = plt.subplots(3, 1, figsize=(10, 7))

axs[0].plot(data[data['labelFreeStyle'] == 1]['gyroscope_X (rad/s)'], label='free style')
axs[0].plot(data[data['labelBreastStroke'] == 1]['gyroscope_X (rad/s)'], label='breast stroke')
axs[0].plot(data[data['labelBackStroke'] == 1]['gyroscope_X (rad/s)'], label='back stroke')

axs[1].plot(data[data['labelFreeStyle'] == 1]['gyroscope_Y (rad/s)'], label='free style')
axs[1].plot(data[data['labelBreastStroke'] == 1]['gyroscope_Y (rad/s)'], label='breast stroke')
axs[1].plot(data[data['labelBackStroke'] == 1]['gyroscope_Y (rad/s)'], label='back stroke')

axs[2].plot(data[data['labelFreeStyle'] == 1]['gyroscope_Z (rad/s)'], label='free style')
axs[2].plot(data[data['labelBreastStroke'] == 1]['gyroscope_Z (rad/s)'], label='breast stroke')
axs[2].plot(data[data['labelBackStroke'] == 1]['gyroscope_Z (rad/s)'], label='back stroke')

axs[0].set_title('Gyroscope')
axs[0].set_ylabel('X (rad/s)')
axs[1].set_ylabel('Y (rad/s)')
axs[2].set_ylabel('Z (rad/s)')
axs[2].set_xlabel('Time (ms)')
axs[0].legend()
axs[1].legend()
axs[2].legend()
```

```python
# plot accelerometer x, y, z data for all styles in three subfigures

fig, axs = plt.subplots(3, 1, figsize=(10, 10))
axs[0].plot(data[data['labelFreeStyle'] == 1]['accelerometer_X (m/s^2)'], label='free style')
axs[0].plot(data[data['labelBreastStroke'] == 1]['accelerometer_X (m/s^2)'], label='breast stroke')
axs[0].plot(data[data['labelBackStroke'] == 1]['accelerometer_X (m/s^2)'], label='back stroke')

axs[1].plot(data[data['labelFreeStyle'] == 1]['accelerometer_Y (m/s^2)'], label='free style')
axs[1].plot(data[data['labelBreastStroke'] == 1]['accelerometer_Y (m/s^2)'], label='breast stroke')
axs[1].plot(data[data['labelBackStroke'] == 1]['accelerometer_Y (m/s^2)'], label='back stroke')

axs[2].plot(data[data['labelFreeStyle'] == 1]['accelerometer_Z (m/s^2)'], label='free style')
axs[2].plot(data[data['labelBreastStroke'] == 1]['accelerometer_Z (m/s^2)'], label='breast stroke')
axs[2].plot(data[data['labelBackStroke'] == 1]['accelerometer_Z (m/s^2)'], label='back stroke')

axs[0].set_title('Linear Accelerometer')
axs[0].set_ylabel('X (m/s^2)')
axs[1].set_ylabel('Y (m/s^2)')
axs[2].set_ylabel('Z (m/s^2)')
axs[2].set_xlabel('Time (ms)')
axs[0].legend()
axs[1].legend()
axs[2].legend()
```

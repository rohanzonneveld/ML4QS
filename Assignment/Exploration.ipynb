```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# load data
acc = pd.read_csv('data/phone/20230608/Linear Accelerometer.csv')
gyro = pd.read_csv('data/phone/20230608/Gyroscope.csv')
baro = pd.read_csv('data/phone/20230608/Barometer.csv')
loc = pd.read_csv('data/phone/20230608/Location.csv')

print("acc: ", len(acc))
print("gyro: ", len(gyro))
print("baro" , len(baro))
print("loc ", len(loc))
```

```python
_baro = baro.loc[808:816]
baroplot = _baro.plot(x='Time (s)', title="Barometer", figsize=(20, 5))
baroplot.set_ylabel("hectoPascal (hPa)")

_baro['Time (s)']
# The start of the first swimming technique is identified at index 814 which represents the 771 second of the data
```

```python
_baro = baro.loc[1150:1180]
baroplot = _baro.plot(x='Time (s)', title="Barometer", figsize=(20, 5))
baroplot.set_ylabel("hectoPascal (hPa)")

_baro['Time (s)']
# The start of the first swimming technique is identified at index 814 which represents the 771 second of the data
```

```python
# plot gyroscope data
free_style = slice(814, 1125)
breast_stroke = slice(1125, 1400)
back_stroke = slice(1400, 1750)

fig, axs = plt.subplots(3, 1, figsize=(10, 10))
axs[0].plot(gyro['X (rad/s)'][free_style], label='free style')
axs[0].plot(gyro['X (rad/s)'][breast_stroke], label='breast stroke')
axs[0].plot(gyro['X (rad/s)'][back_stroke], label='back stroke')

axs[1].plot(gyro['Y (rad/s)'][free_style], label='free style')
axs[1].plot(gyro['Y (rad/s)'][breast_stroke], label='breast stroke')
axs[1].plot(gyro['Y (rad/s)'][back_stroke], label='back stroke')

axs[2].plot(gyro['Z (rad/s)'][free_style], label='free style')
axs[2].plot(gyro['Z (rad/s)'][breast_stroke], label='breast stroke')
axs[2].plot(gyro['Z (rad/s)'][back_stroke], label='back stroke')

axs[0].set_title('Gyroscope')
axs[0].set_ylabel('X (rad/s)')
axs[1].set_ylabel('Y (rad/s)')
axs[2].set_ylabel('Z (rad/s)')
axs[2].set_xlabel('Time (ms)')
axs[0].legend()
axs[1].legend()
axs[2].legend()
```
